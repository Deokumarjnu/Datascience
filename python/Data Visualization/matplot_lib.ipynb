## MatplotLib

```python
import matplotlib.pyplot as plt
%matplotlib inline #used to not write plt.show() ever time
```

```python
import numpy as np
```

```python
# arange is used for creating arrays
#examles
x = np.arange(0 ,10) 
y = np.arange(10 ,20) 
```

### Scatter plot

```python
#lets plot simple scatter plot
plt.scatter(x, y, c = "g")
```

```python
# add title
plt.scatter(x, y, c = "g")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Graph in 2D")
```

### Plot

```python
# plot
y = x*x
plt.plot(x, y, "r*",linestyle='dashed', linewidth=2, markersize=12)
# plt.plot(x, y, "r"/"--r"/"--ro"/"r*", )
plt.xlabel("X Axis")
plt.xlabel("Y Axis")
plt.title("2D Graph")
```

```python
# creating subplot
plt.subplot(2,2,1) #(number of rows, num of columns, position)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'g*', linestyle="dashed")
plt.subplot(2,2,3)
plt.plot(x,y,'bo', linestyle="dotted")
plt.subplot(2,2,4)
plt.plot(x,y,'go', linestyle="solid")
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine wave form")
plt.plot(x, y)
```

```python
# lets build sin and cos diagram use of subplot
x = np.arange(0, 4*np.pi, 0.1)
y_sin= np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,2,1)
plt.plot(x, y_sin,"g--")
plt.title("Sine wave form")

plt.subplot(2,2,2)
plt.plot(x, y_cos,"r--")
plt.title("Cos wave form")
```

### Bar plot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x,y)
plt.bar(x2,y2, color="g")

plt.title("Bar plot")
plt.xlabel("X Axix")
plt.ylabel("Y Axix")
```

### Histogram

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a)
plt.title("Histogram") #x - axix there will be all the values and counts will be on y axix
```

### Box plot - mainly used for find percentile and outliers

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

plt.boxplot(data, vert=True, patch_artist=True) # patch_artist-  is used to filed the color, vert - is used for horzontal or verticle graph
```

### Pie Chart 

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4, 0, 0, 0)  # explode 1st slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%', shadow=False)
```
