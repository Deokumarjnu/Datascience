```python
# turn the item price into a float
float_num = lambda x : float(x[1:-1])
df.item_price = df.item_price.apply(float_num)
```

```python
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv"
df = pd.read_csv(url)
df.head(5)
```

```python
# convert the year to datetime64
df.Year = pd.to_datetime(df.Year, format="%Y")
df.info()
```

```python
# set Year column as index of dataframe
df = df.set_index(df.Year, drop=True)
df
```

```python
# Group the year by decades and sum the values
# 1. uses resample to sum each decade
crime = df.resample('10AS').sum()
# 2. uses resample to get the max value only for the "Population" column
population = crime["Population"].resample("10AS").max()
#  3. Updating the "Population" column

crime["Population"] = population
crime
```

```python
df = crime
```

```python
# return the first 3 of df
df.iloc[:3]
```

```python
# select just the Murder and Robbery columns from the Dataframe df and print
df.loc[ : , ['Murder','Robbery']].head()
```

```python
# select the data in rows [3,4,8] and in columns ["Murder","Robbery"]
df.loc[df.index[[3, 4, 8]], ['Murder', 'Robbery']]
```

```python
# select only the rows where the number of murder is greater than 24,000
df[df.Murder > 24000]
```

```python
# select the rows the murder is between 20k and 24k
df[df['Murder'].between(20000, 24000)]
```

```python
# Calculate the mean murder for each different year in df
df.groupby("Year")["Murder"].mean()
```

```python
# Sort df first by the values in the Murder in decending order then by the value in the Violent column in ascending order
df.sort_values(by=['Murder', 'Violent'], ascending=[False, True])
```
