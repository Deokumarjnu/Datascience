```python
import requests
```

```python
url = "https://api.github.com/repos/pandas-dev/pandas/issues"
response = requests.get(url)
type(response)
```

```python
data = response.json()
```

```python
#Extract -> url, comments_url, html_url, user (note-Unwrappe the labels and user in same label of datafram)
import pandas as pd
df = pd.DataFrame(data, columns=["url","comments_url","html_url","user"])
df
```

```python
# Split keys and values in columns
from pandas.io.json import json_normalize

user_data = json_normalize(df["user"])
user_data.head()

# Concatinating both the data sets
conct_df = pd.concat([df[:], user_data[:]], axis=1)
conct_df.head()

#remove/unselect user column since it's already splited above
dataFrame = pd.concat([df[["url","comments_url","html_url"]], user_data[:]],axis=1)
dataFrame
```
