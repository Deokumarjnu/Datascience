### Univariate Analysis - will take one feature and do the analysis

```python
data_setosa = data.loc[data["Species"]== "setosa"]
data_virginica = data.loc[data["Species"]== "virginica"]
data_versicolor = data.loc[data["Species"]== "versicolor"]
```

```python
# plt.plot(x,y) - will plot it for sepal length
plt.plot(data_setosa['sepalLengthCm'], np.zeros_like(data_setosa['sepalLengthCm']), 'o')
plt.plot(data_virginica['sepalLengthCm'], np.zeros_like(data_virginica['sepalLengthCm']), 'o')
plt.plot(data_versicolor['sepalLengthCm'], np.zeros_like(data_versicolor['sepalLengthCm']), 'o')
plt.show()
```

#### So in the above graph we can clearly say tht the max lenth of setosa is arroun ~4.8, virginica will ~4.4 - ~7.3 and vericolor will be above ~7.4

### Bivariate Analysis - will take two feature

```python
# will use FacetGrid function of seaborn library to visualize the classification of iris data for sepalLengthCm and sepalWidthCm
sns.FacetGrid(data, hue="Species",size=5).map(plt.scatter,"sepalLengthCm","SepalWidthCm").add_legend()
```

#### From the above grap we can see the setosa is classified separetly but versicolor and virginica are mixed , so this plot will help us to determine which algorithms should we use for model building wether linear (regration) or non-linear(DT, RF, SVM, KNN etc)

### Multivariate Analysis - used for more than 3 features

```python
# Will use pairplot to see the different kind of relationship between variables
sns.pairplot(data, hue="Species", size=3)
```
